```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.metrics import classification_report, mean_squared_error as mse, r2_score as r2
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
```

# Preprocessing

```python
x_df = pd.read_csv(r'C:\Training\Demo\Python\pytorch training\ml mini project\datasets\winequality_red_x_train.csv')
y_df = pd.read_csv(r'C:\Training\Demo\Python\pytorch training\ml mini project\datasets\winequality_red_y_train.csv')
```

```python
def reverse_ohe(row):
    if row['3'] == 1:
        return 3
    elif row['4'] == 1:
        return 4
    elif row['5'] == 1:
        return 5
    elif row['6'] == 1:
        return 6
    elif row['7'] == 1:
        return 7
    elif row['8'] == 1:
        return 8
y_df['label'] = y_df.apply(reverse_ohe, axis=1)
y_df
```

```python
#split training and testing set 
x_train, x_test, y_train, y_test = train_test_split(x_df, y_df['label'], test_size=0.2, random_state=0)
```

```python
y_train.value_counts()
```

# Model Selection

```python
models = [
    LogisticRegression(), 
    MLPClassifier(), 
    KNeighborsClassifier(), 
    SVC(),
    GaussianProcessClassifier(), 
    QuadraticDiscriminantAnalysis(), 
    DecisionTreeClassifier(), 
    RandomForestClassifier(), 
    AdaBoostClassifier(), 
    GaussianNB()
] 
```

```python
for model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    print('\n\n\n' + str(model))
    #show the results in a confusion matrix
    # cf_matrix = confusion_matrix(y_test, y_pred)

    # disp_cf = sns.heatmap(cf_matrix, annot=True, cmap='Blues', fmt='d')

    # disp_cf.set_title('Confusion Matrix')
    # disp_cf.set_xlabel('\nPredicted Category')
    # disp_cf.set_ylabel('Actual Category ')
    # plt.show()

    print(classification_report(y_test, y_pred))
    
```
